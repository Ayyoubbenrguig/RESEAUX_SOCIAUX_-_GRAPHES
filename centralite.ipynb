```python
import networkx as nx
import matplotlib.pyplot as plt

def reseau(network_name, edges):
    # Création du graphe
    G = nx.Graph()

    # Ajout des arcs à coûts unitaires
    G.add_edges_from(edges, weight=1)

    # Affichage du graphe avec les noeuds en rouge
    pos = nx.spring_layout(G, seed=500)
    nx.draw(G, pos, with_labels=True, node_size=500, node_color='red')  # Set node_color to 'red'
    plt.title(f"{network_name}", fontsize=20)
    plt.rcParams['figure.figsize'] = 15, 8
    plt.show()

    return G
```

### **b. Discussion des effets de $\alpha$ et de $\beta$ pour réseau_0**

L'importance propre d'un noeud est relié à la valeur de $\beta$ et l'importance liée à ses voisins est reliée à la valeur de $\alpha$.

**Cas 1**

 Lorsque $\beta$ $= 0 $  ,la mesure de centralité ne parvient pas à attribuer des valeurs significatives aux nœuds et donc elles deviennent toutes nulles et ce quelque soit la valeur de $\alpha$.


 **Cas 2**

 Lorsque $\alpha$ $ \lt 1$ et élevé $(0,85)$ ,  les nœuds très connectés ou avec de nombreux liens sortants voient leur centralité augmenter. Ces nœuds deviennent des points focaux importants dans le réseau.
Les nœuds moins connectés ou isolés voient leur centralité diminuer, car le surfeur aléatoire a tendance à rester concentré autour des nœuds fortement connectés.


**Cas 3**

 Lorsque $\alpha$ $ =0 $ et $\beta$ $= 1 $  , la centralité obtenue est liée uniquement à $\beta$ qui est le terme de biais et ainsi la centralité de chaque noeud reste celle qui lui est propre donc une valeur de $1$ pour chaque noeud.

### **c) Interprétation**:

Dans le cadre du PageRank il faut choisir des valeurs de alpha inférieure à $1/$ $\lambda$, $\lambda$ étant la plus grande valeur propre de la matrice d'adjacence du réseau, pour s'assurer de la convergence des centralités.Cependant lorsque $\alpha$ est encore plus proche de 1 la convergence s'avère rapide.



### **d) Ecrire un algorithme pour implémenter le « PageRank » pour n'importe quel graphe. L'algorithme prend en entrée graphe, α et β, et fournit en sortie les valeurs « PageRank »**


```python
# Normalisation de la centralité du Pagerank pour l'utilisation de la Methode des Puissances
def calculate_and_normalize_pagerank(graph, alpha, beta, tol=1e-6, max_iter=100):
    # Convert the graph to a NetworkX Graph object
    G = nx.Graph(graph)

    # Adjacency matrix
    A = nx.to_numpy_array(G)

    # Number of nodes in the graph
    n = len(G)

    # Initialize the PageRank vector
    pagerank_vector = np.ones(n) / n

    # Power method iteration
    for _ in range(max_iter):
        # Update the PageRank vector without damping
        new_pagerank_vector = alpha * A @ pagerank_vector + beta

        # Check for convergence
        if np.linalg.norm(new_pagerank_vector - pagerank_vector, 1) < tol:
            break

        pagerank_vector = new_pagerank_vector

    # Normalize PageRank values
    total_pagerank = sum(pagerank_vector)
    normalized_pagerank = {node: score / total_pagerank for node, score in zip(G.nodes, pagerank_vector)}

    return normalized_pagerank
```

```python
# Algorithme du page rank 
def page_rank(graph, alpha, beta):
    print("pour alpha =", alpha, "et pour beta =", beta)
    nodes = list(graph.keys())
    num_nodes = len(nodes)

    A = np.zeros((num_nodes, num_nodes))
    for i, node in enumerate(nodes):
        for neighbor in graph[node]:
            j = nodes.index(neighbor)
            A[i, j] = 1
    # print("la matrice A est:")
    # print(A)

    D = np.diag([len(graph[node]) for node in nodes])
    # print("la matrice D est:")
    # print(D)

    # Création d'une matrice colonne remplie de 1
    matrice_colonne = np.ones((num_nodes, 1))
    
    inverse_D = np.linalg.inv(D)
    I = np.eye(num_nodes)
    transposee_A = np.transpose(A)
    J = np.dot(transposee_A, inverse_D)
    W = I - alpha * J
    Z = np.linalg.inv(W)
    K = np.dot(Z, matrice_colonne)
    C = beta * K

    # Création du dictionnaire de résultats
    resultats_pagerank_neighborhood = {nodes[i]: C[i, 0] for i in range(num_nodes)}

    # Tri des nœuds par score décroissant
    sorted_nodes = sorted(resultats_pagerank_neighborhood, key=resultats_pagerank_neighborhood.get, reverse=True)

    # Affichage du classement
    if beta != 0:       
        print("Classement des noeuds par score de centralite d'importance du voisinage:")
        for rank, node in enumerate(sorted_nodes, start=1):
            print(f"{rank}. Noeud {node}: PageRank = {resultats_pagerank_neighborhood[node]}")
        return C
    else:
        # Calculate and normalize PageRank with alpha=1 and beta=0
        normalized_pagerank = calculate_and_normalize_pagerank(graph, alpha, beta)
        sorted_nodes = sorted(normalized_pagerank, key=normalized_pagerank.get, reverse=True)

        # Display normalized PageRank values
        print("Classement des noeuds par score de centralite d'importance suivant la methode des puissances avec 100 iterations :")
        # for node, score in normalized_pagerank.items():
        #     print(f"Node {node}: Normalized PageRank = {score}" )
        for rank, node in enumerate(sorted_nodes, start=1):
            print(f"{rank}. Noeud {node}: PageRank = {normalized_pagerank[node]}")
            
```

### **a. Ecrire un algorithme pour calculer le nombre de plus court chemin entre deux nœuds :**

```python
def nombre_plus_courts_chemins(graph, s, t):
    try:
        # Calcul des plus courts chemins de s à t
        all_shortest_paths = list(nx.all_shortest_paths(graph, source=s, target=t))
        return len(all_shortest_paths)
    except nx.NetworkXNoPath:  # Si aucun chemin n'est trouvé
        return 0  # Il n'existe pas de chemin entre s et t
    
s = 'A'
t = 'I'

print(f"Le nombre de court chemin entre {s} et {t} est", nombre_plus_courts_chemins(reseau_0, s, t))
```

### **b. Ecrire un algorithme pour calculer le nombre de plus court chemin entre deux nœuds et qui passe par un nœud $v$, donné. Attention le résultat peut être zéro, sinon aucun cours chemin entre ces deux nœuds ne passe par $v$**

```python
def nombre_plus_courts_chemins_passant_par_v(graph, s, t, v):
    try:
        # Calcul des plus courts chemins de s à t
        all_shortest_paths = list(nx.all_shortest_paths(graph, source=s, target=t))

        # Filtrer les chemins passant par le nœud intermédiaire v
        paths_passant_par_v = [path for path in all_shortest_paths if v in path]

        return len(paths_passant_par_v)
    except nx.NetworkXNoPath:  # Si aucun chemin n'est trouvé
        return 0  # Il n'existe pas de chemin entre s et t

s = 'A'
t = 'I'
v = 'F'

print(f"Le nombre de court chemin entre {s} et {t} passant par {v} est", nombre_plus_courts_chemins_passant_par_v(reseau_0, s, t, v))
```

### **c. Ecrire un algorithme pour implémenter la centralité d’intermédiarité pour n'importe quel graphe.**

```python
def centralite_intermediarite(graph):
    # Create a graph using reseau_0
    graph = nx.Graph(graph)
    nodes = list(graph.nodes)
    nombre_noeuds = len(nodes)

    # Initialiser les valeurs de centralité à 0.0 pour tous les nœuds
    valeurs_centralite = {node: 0.0 for node in nodes}

    for node in nodes:
        for source in nodes:
            if node != source:
                for target in nodes:
                    if node != target and source != target:
                        nombre_plus_courts_chemins_st_v = nombre_plus_courts_chemins_passant_par_v(graph, source, target, node)
                        nombre_plus_courts_chemins_st = nombre_plus_courts_chemins(graph, source, target)
                        valeurs_centralite[node] += nombre_plus_courts_chemins_st_v / nombre_plus_courts_chemins_st

    # Facteur de normalisation
    facteur_normalisation = (nombre_noeuds - 1) * (nombre_noeuds - 2)

    # Normaliser les valeurs de centralité
    for node in nodes:
        valeurs_centralite[node] /= facteur_normalisation

    return valeurs_centralite
```

## **3) Centralité de proximité**

### **a. Ecrire un algorithme pour calculer le plus court chemin entre un nœud et les autres et retourne la moyenne de ces chemins.**

```python
def shortest_paths(graph, start_node):
    # Initialisation des distances
    distances = {node: float('inf') for node in graph.nodes()}  # Initialise les distances à l'infini pour tous les nœuds du graphe
    distances[start_node] = 0  # La distance du nœud de départ à lui-même est mise à 0
    queue = [(0, start_node)]  # Crée une file de priorité avec le nœud de départ et sa distance (0)

    # Parcours en largeur (BFS)
    while queue:  # Tant que la file de priorité n'est pas vide
        current_distance, current_node = queue.pop(0)  # Prend le premier élément de la file (plus petite distance actuelle)
        for neighbor in graph.neighbors(current_node):  # Pour chaque voisin du nœud actuel dans le graphe
            new_distance = current_distance + 1  # Calcule la nouvelle distance, avec un poids de 1 pour chaque arête
            if new_distance < distances[neighbor]:  # Vérifie si cette nouvelle distance est plus courte que la connue précédemment
                distances[neighbor] = new_distance  # Met à jour la distance vers le voisin
                queue.append((new_distance, neighbor))  # Ajoute le voisin à la file avec sa nouvelle distance

    return distances  # Retourne un dictionnaire contenant les distances les plus courtes depuis le nœud de départ vers tous les autres nœuds du graphe
```

### **b. Ecrire un algorithme pour implémenter la centralité proximité pour n'importe quel graphe.**

```python
def centralite_proximite(graph):
    # Convert the dictionary to a NetworkX graph
    graph = nx.Graph(graph)
    closeness_centralities = {}
    
    for node in graph.nodes():
        distances = shortest_paths(graph, node)
        reachable = sum(1 for dist in distances.values() if dist != float('inf'))
        total_distance = sum(dist for dist in distances.values() if dist != float('inf'))

        if total_distance > 0:
            closeness_centralities[node] = (reachable - 1) / total_distance
            closeness_centralities[node] *= (reachable - 1) / (len(graph.nodes()) - 1)
        else:
            closeness_centralities[node] = 0

    return {k: v for k, v in sorted(closeness_centralities.items(), key=lambda item: item[1], reverse=True)}
```

```python
def creer_graphe_a_partir_des_aretes(nom_fichier):
    with open(nom_fichier, 'r') as fichier:
        lignes = fichier.readlines()
        nombre_noeuds, nombre_aretes = map(int, lignes[0].split())
        aretes = [tuple(map(str, ligne.split())) for ligne in lignes[1:]]
    
    G = nx.Graph()
    G.add_nodes_from(map(str, range(1, nombre_noeuds + 1)))  # Make nodes strings

    # Create a graph with the same structure as reseau_0
    graph_dict = {}
    for edge in aretes:
        node = str(edge[0])
        neighbor = str(edge[1])
        if node not in graph_dict:
            graph_dict[node] = [neighbor]
        else:
            graph_dict[node].append(neighbor)

        # Ensure bidirectional edges
        if neighbor not in graph_dict:
            graph_dict[neighbor] = [node]
        else:
            graph_dict[neighbor].append(node)

    # Order the graph dictionary by keys
    ordered_graph_dict = dict(sorted(graph_dict.items(), key=lambda x: int(x[0])))
    
    return ordered_graph_dict                   # Retourne la liste complète des arêtes extraites du fichier
```

```python
def edges(file_name):
    edges = []           # Initialise une liste vide pour stocker les arêtes
    with open(file_name, 'r') as file:               # Ouvre le fichier en mode lecture
        lines = file.readlines()                 # Lit toutes les lignes du fichier
        num_nodes, num_edges = map(int, lines[0].split())              # Extrait le nombre de nœuds et d'arêtes à partir de la première ligne
        for line in lines[1:]:                    # Parcourt les lignes restantes du fichier
            edge = tuple(line.split())                 # Divise la ligne en une paire d'éléments et stocke l'arête comme un tuple
            edges.append(edge)                  # Ajoute l'arête à la liste des arêtes
    return edges                    # Retourne la liste complète des arêtes extraites du fichier
```
